```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df
```

```python
df_sorted = df.sort_values('Age')
ax = df_sorted.groupby('Age').size().plot(kind='bar', figsize=(12, 6), title='Purchase Distribution by Age')
plt.xticks(rotation=0)
plt.show()
```

```python
lst = []
for i in df['Age'].unique():
    lst.append([i , df[df['Age']==i]['Product_ID'].nunique()])


data = pd.DataFrame(lst ,  columns = ['Age', 'No of Products'])
```

```python
data
```

```python

ax = data.plot.bar(x = 'Age', figsize = (6,6))
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('Marital_Status').size()
```

```python
ax = df.groupby('Age')['Purchase'].sum().plot(kind = 'bar', figsize = (8, 6), title = 'Amount Spend by Age')
plt.xticks(rotation=0)
plt.show()
```

```python
ax = df.groupby('Age')['Purchase'].mean().plot(kind = 'pie', figsize = (6, 6), title = 'Amount Spend by Age' , autopct = "%0.2f")
plt.xticks(rotation=0)
plt.show()
```

```python
df.groupby('Marital_Status').size().plot(kind = 'pie',figsize=(6,6) , autopct = '%0.2f' , title='Married Percentage')
```
