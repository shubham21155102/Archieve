```python
import pandas as pd
```

```python
df = pd.read_csv('BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
for column in df.columns:
    print(df[column].nunique(), "\t: ", column)
```

```python
data = pd.DataFrame({'Ratio' : [len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, 
                    index = ['Male', 'Female'])

# data

data.plot(kind='pie' ,y = 'Ratio', figsize = (6,6), autopct = "%.1f")
```

## Performing similar operation in single line without using logic 

```python
df.groupby('Gender').size()
```

```python
df.groupby('Gender').size().plot(kind = 'pie', 
                                autopct = "%.2f",
                                title = 'Gender Ratio',
                                figsize = (4,4))
```

```python
# df.groupby('Gender').size().plot(kind = 'bar', 
#                                 figsize = (4,4))

from matplotlib import pyplot as plt

ax = df.groupby('Gender').size().plot(kind='bar' , figsize = (4,4))
plt.xticks(rotation=0)
plt.show()


```
