```python
from matplotlib import pyplot as plt
gender_ratio_df = df.pivot_table(index='Occupation', columns='Gender', aggfunc='size', fill_value=0)

# Calculate the male to female ratio
df['Male_to_Female_Ratio'] = gender_ratio_df['M'] / gender_ratio_df['F']

plt.plot(df['Male_to_Female_Ratio'])
```
